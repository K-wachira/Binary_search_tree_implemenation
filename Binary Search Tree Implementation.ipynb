# Binary Search Tree Implementations

```python
# Definition of a binary Tree Node

class TreeNode():
    def __init__(self, val = 0, left = None, right= None):
        self.val = val
        self.left = left
        self.right = right
```

```python
# Implementation of a binary Tree
class Solution(object):
    def __init__ (self):
        
        self.root = None
        
#      Inserts a given value in a Tree
    def insertion(self, value): 
        if self.root == None: # if there is no Tree in the instance called  initializes with root being the value inserted first
            self.root = TreeNode(value)
        else:            # if a tree already exists call the private insertion method to deal with the logic  // I have split the methods for ease in understanding and reading the code
            self._insertion( self.root, value)
            
    def _insertion(self, curr_node, value): # This is a private method that will handle the recursion and the insertion logic 
        
        if value < curr_node.val:       # if its less than we will be going left side
            if curr_node.left == None:  # checks if the left node exists and if it does not create one and insert the value on the new node
                curr_node.left = TreeNode(value) 
            else:                       # if the left node had a value already i.e not null make the left node the parent while recursing 
                self._insertion(curr_node.left, value)
                
        elif value > curr_node.val:     # if the value we are inserting is greater than the parent go right
            if curr_node.right == None: # if there is no right Node to go to create a Node left of the parent and insert the value as the val
                curr_node.right= TreeNode(value)
            else:                         # If there is a node on the right side , make it the parent when recursing
                self._insertion(curr_node.right, value)
        else: 
            print("Value is already on the Tree") # this is to avoid dublication of values on a tree
            
            
            
#      checks if a given value is in Tree       
    def search(self, value):
        if self.root == None:   # First check if the Tree is empty 
            print("The Tree is currently Empty")
        else:
            return self._search(self.root, value)  #If tree not Empty return the private methodsreturn value 
    def _search(self, curr_node, value):
        if curr_node == None:         # Checks if we ran out of Nodes
            return 'Value Not Found'
        else:
            if curr_node.val == value:  # Acts as the base case when we find the value
                return "Value " + str(value) + " is in Tree"
            
            elif value  < curr_node.val:   #if the value is less than the parents value we should go left
                return self._search(curr_node.left, value)  # we are returning because we want to ONLY consider children and grandchildren of the left Node
            elif value > curr_node.val:    # if the value is more than the parents value we should focus on the right side of the parent as the value is more likely to be there
                return self.search (curr_node.right, value)  # we are returning bacause we only want to consider the children of the sub-tree on the right side of the current node(parent)
            
            
    def get_height(self):
        if self.root == None:  # Check if root is empty 
            return "The Tree is currently Empty hence height = 0"
        else:
            return self._get_height(self.root, 0)  
            
    def _get_height(self, curr_node, height):
        if curr_node == None: return height
        left = self._get_height(curr_node.left, height+1)  
        right = self._get_height(curr_node.right, height+1)
        
        return max(left, right)
            
            

            # Root Left Right
    def printing_pre_order(self):
        if self.root == None:
            print("The Tree is currently Empty")
        else:
            self._printing_pre_order(self.root)
            
    def _printing_pre_order(self, curr_node):
        if curr_node != None:
            print(curr_node.val)   
            self._printing(curr_node.left)

            self._printing(curr_node.right)
            
            #  Left Root Right
    def printing_in_order(self):
        if self.root == None:
            print("The Tree is currently Empty")
        else:
            self._printing_in_order(self.root)
            
    def _printing_in_order(self, curr_node):
        if curr_node != None:
            self._printing(curr_node.left)
            print(curr_node.val)   
            self._printing(curr_node.right)
            
            
            #  Left Root Right
    def printing_post_order(self):
        if self.root == None:
            print("The Tree is currently Empty")
        else:
            self._printing_post_order(self.root)
            
    def _printing_post_order(self, curr_node):
        if curr_node != None:
            self._printing(curr_node.left)
            self._printing(curr_node.right)
            print(curr_node.val)   
            
            
            
            
            


Tree1 = Solution()
Tree1.insertion(10)
Tree1.insertion(11)
Tree1.insertion(9)
Tree1.printing()
```
